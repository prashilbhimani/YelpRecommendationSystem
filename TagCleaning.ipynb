```python
import json
import nltk
```

```python
business_tags={}
for b in open('az_business_tags.json','r', encoding="utf8"):
    business=json.loads(b)
    business_tags[business['business_id']]={}
```

```python
len(business_tags.keys())
```

```python
j=0
for b in business_tags.keys():
    temp={}
    og=len(business_tags[b])
    if og>0:
        for k in business_tags[b].keys():
            if business_tags[b][k] is not None:
                k1=nltk.stem.WordNetLemmatizer().lemmatize(k,'n')
                if k1 in temp and temp[k1] is not None:
                    temp[k1]=temp[k1].append(business_tags[b][k])
                else:
                    temp[k1]=business_tags[b][k]
    business_tags[b]=temp
    

        
```

```python
sum_tags=0
for b in business_tags.keys():
    sum_tags=sum_tags+len(business_tags[b].keys())
print(sum_tags/len(business_tags.keys()))
```

```python
with open("Dataset/JSON_Dataset/az_business_tags.json","w") as outfile:
    json.dump(business_tags,outfile)
    outfile.write("\n")
```
